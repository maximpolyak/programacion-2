```python
import requests
from bs4 import BeautifulSoup as bs 

url = "https://www.passiton.com/inspirational-quotes"
rq = requests.get(url=url)

#Traemos el contenido de la consulta con request, y le indicamos que la vamos a analizar con la librería de html5lib. Prettify lo muestra más amigablemente, según los datos.
soup = bs(rq.content, "html5lib")
#print(soup)
soup_pt = soup.prettify()
print(soup_pt)
```

```python
#En este ejemplo queremos recoger las citas, el tema y la url de cada cita

#crear una lista para alcenar los resultados
results = []

#recoger todos los elementos dentro del contenedor que nos interesa y guardar cada uno en un diccionario
div_citas = soup.find("div", attrs = { "id":"all_quotes" })

for elem in div_citas.find_all("div"):
    result = {}
    result["theme"] = elem.h5.text
    result["url"] = elem.a["href"]
    result["lines"] = elem.img["alt"].split("#")[0]
    results.append(result)

print(results)
```

```python
#Guardar los resultados en un archivo .csv -> información para análisis posteriores

import csv

filetosave = "data/frases_inspiradoras.csv"
with open(filetosave,"w",newline="") as file:
    add = csv.DictWriter(file, ["theme","url","lines"])
    add.writeheader()
    for result in results:
        add.writerow(result)
```

```python
#Guardar un fichero por consulta -> usar para alimentar ccia datos

from datetime import datetime
import csv

now = datetime.now()
now_str = now.strftime("%Y%m%d%H%M%S")

filetosave = "data/" + now_str + "_frases_inspiradoras.csv"
with open(filetosave,"w",newline="") as file:
    add = csv.DictWriter(file, ["theme","url","lines"])
    add.writeheader()
    for result in results:
        add.writerow(result)
```
